```python
# libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
df = pd.read_csv('data/korean-top-youtube.csv')
df.iloc[:] = df.iloc[::-1].values # reverse the df, without change its index
df['Akun Youtube'] = df['Akun Youtube'].replace({'1theK (원더케이)':'1theK'}) # remove foreign language from '1theK'
df['Jumlah Subscriber'] = df['Jumlah Subscriber'].str.replace(r'M','') # remove the 'M' from 'Jumlah Subscriber'
df['Jumlah Subscriber'] = df['Jumlah Subscriber'].astype(float)
df.head()
```

```python
extra = np.arange(0,10) # initiate an extra variable for plotting
```

```python
# Plot
fig = plt.figure(figsize=(17,8))
plt.hlines(y=extra, xmin=0, xmax=df['Jumlah Subscriber'], color='orange', alpha=0.7, linewidth=2)
plt.scatter(df['Jumlah Subscriber'], extra, color='orange')
plt.title('Top 10 Youtube Account in South Korea')
plt.xlabel('Number of Subscribe (M)')
plt.yticks(extra, df['Akun Youtube'])

# Adding some annotations
for row in df.itertuples():
    plt.text(x = row[2]+.1, y = row.Index, s = str(row[2]) + 'M', horizontalalignment= 'center',
             verticalalignment='bottom', fontsize=12)

plt.show()
```

```python
# saving plot
fig.savefig('korean-top-youtube.jpg')
```
